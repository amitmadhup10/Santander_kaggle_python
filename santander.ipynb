```python
def remove_zero_var_pred(train):
    features = []
    for i in train.columns:
        if train[i].var() != 0:
            features.append(i)
    return(features)
```

```python
def remove_highly_correlated(train):
    feature_i = []
    feature_j = []
    for i in range(0,len(train.columns)):
        for j in range(i+1,len(train.columns)):
            if np.corrcoef(train.iloc[:,i],train.iloc[:,j])[0,1] > 0.90:
                feature_i.append(train.columns[i])
                feature_j.append(train.columns[j])
            j= j+1
        i= i+1
    return(pd.DataFrame(feature_i,feature_j))
```
